## Deflexión de la viga

```python
L= 10 #longitud de la viga
F= -15 #fuerza aplicada
E=29000*(12**2) #Módulo elàstico del material
I=758/(12**4) #inercia
x=20
def Deflexión(x,y):
    return (F/(E*I))*((L*(x*x))-((x*x*x)/6))

print(Deflexión(x,y))

x0 = 1
tt = np.linspace(0, L+1)
xx = odeint(Deflexión, x0, tt)

plt.figure(figsize=(15,10))
plt.plot(tt, xx, 'g', label = 'Ecuación de la Viga')
plt.grid()
plt.legend(loc = 'best')
plt.xlabel('Distancia $(ft)$')
plt.ylabel('Deflexión')
plt.title('Deflexión de una viga')
plt.show()
```
